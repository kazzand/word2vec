```python
text = 'one two three four five mouse fuck\n horse cool five man goga\n five four cmok lave\n'
```

```python
import numpy as np
def train(data):
    corpus = data.split('\n')
    
    for i in range(len(corpus)):
        corpus[i] = corpus[i].lower().split() 
    voc = {}
    inv_voc = {}
    index = 0
    for sentence in corpus:
        for word in sentence:
            if word not in voc:
                voc[word] = index
                inv_voc[str(index)] = word
                index += 1
   
    window_size = 2
    training_data = []
    
    vocabulary_size = len(voc)
    embedding_dims = 5
    lr = 0.1
    def get_input(word_idx):
        x = np.zeros(vocabulary_size)
        x[word_idx] = 1.0
        return x
    
    #Создадим пары слов (центральное слово - контекстное слово)
    for sentence in corpus:
        indices = []
        for word in sentence:
            indices.append(voc[word])
        
        for mid_pos in range(len(indices)):
            context = []
            for stride in range(-window_size, window_size + 1):
                    
                    context_pos = mid_pos + stride
                
                    if context_pos < 0 or context_pos >= len(indices) or context_pos == mid_pos:
                        continue
                    
                    context_index = indices[context_pos]
                    context.append(get_input(context_index))
                    
                
            training_data.append((get_input(indices[mid_pos]), context))
      # В pairs в итоге лежат индексы пар контекстно-связанных слов
    def get_input(word_idx):
        x = np.zeros(vocabulary_size)
        x[word_idx] = 1.0
        return x
    
    w1 = np.random.randn(vocabulary_size, embedding_dims)
    w2 = np.random.randn(embedding_dims, vocabulary_size)
    num_epochs = 300

    
    def softmax(x):
        e_x = np.exp(x - np.max(x))
        return e_x / e_x.sum(axis=0)
    
    

    
    for epo in range(num_epochs):
        loss = 0
        for data, target in training_data:
            x = data
            y = target
            
            h = np.dot(w1.T, x)
            u = np.dot(w2.T, h)
            y_pred = softmax(u)
        
            EI = np.sum([np.subtract(y_pred, word) for word in target], axis=0)
            
            dl_dw2 = np.outer(h, EI)  
            dl_dw1 = np.outer(x, np.dot(w2, EI.T))

            # UPDATE WEIGHTS
            w1 -= lr * dl_dw1
            w2 -= lr * dl_dw2
            
            
    w2v_dict = {}
    for word in voc:
        w_index = voc[word]
        w2v_dict[word] = w1[w_index]
        

    return w2v_dict
```

```python
train(text)
```
